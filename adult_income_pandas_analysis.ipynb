```python
# Loading the dataset
data = pd.read_csv("adult.data.csv")
data.head()
```

**1. Calculate how many men and women (feature *sex*) are represented in this dataset**

```python
sex_counts = data["sex"].value_counts()

print(sex_counts)
```

**2. What is the average age of men (feature *age*) across the entire dataset?**

```python
average_age_men = data[data["sex"] == "Male"]["age"].mean()

print(average_age_men)
```

**3. What is the proportion of United States citizens (*native-country* attribute)?**

```python
us_proportion = (data["native-country"] == "United-States").mean() * 100

print(us_proportion)
```

**4-5. Calculate the mean and standard deviation of the age of those who earn more than 50K per year (*salary* attribute) and those who earn less than 50K per year**

```python
# Age statistics for people earning >50K
mean_age_high = data[data["salary"] == ">50K"]["age"].mean()
std_age_high = data[data["salary"] == ">50K"]["age"].std()

# Age statistics for people earning <=50K
mean_age_low = data[data["salary"] == "<=50K"]["age"].mean()
std_age_low = data[data["salary"] == "<=50K"]["age"].std()

print("Mean age (>50K):", mean_age_high)
print("Std age (>50K):", std_age_high)

print("Mean age (<=50K):", mean_age_low)
print("Std age (<=50K):", std_age_low)
```

**6. Is it true that people who earn more than 50k have at least a higher education? (*education indicator – Bachelor's, Professional school, Assoc-acdm, Assoc-voc, Master's* or *Doctorate*)**

```python
higher_education = [
    "Bachelors",
    "Prof-school",
    "Assoc-acdm",
    "Assoc-voc",
    "Masters",
    "Doctorate"
]

high_salary = data[data["salary"] == ">50K"]

result = high_salary["education"].isin(higher_education).all()

print(result)
```

**7. Find the age statistics for each race (*race* attribute) and each gender. Use *groupby* and *describe*. Find the maximum age of males of the *Asian-Pac-Islander* race.**

```python
age_stats = data.groupby(["race", "sex"])["age"].describe()

print(age_stats)
```

```python
max_age_api_male = data[(data["race"] == "Asian-Pac-Islander") & (data["sex"] == "Male")]["age"].max()

print(max_age_api_male)
```

**8. Among whom is the proportion of high-earning men (>50K) higher: married or single men (*marital-status* indicator)? We consider those whose *marital-status* begins with *Married* (Married-civ-spouse, Married-spouse-absent, or Married-AF-spouse) to be married; the rest are considered single.**

```python
# Filter only men
men = data[data["sex"] == "Male"]

# Married men
married = men[men["marital-status"].str.startswith("Married")]

# Single men
single = men[~men["marital-status"].str.startswith("Married")]

# Proportion earning >50K
married_high_income = (married["salary"] == ">50K").mean()
single_high_income = (single["salary"] == ">50K").mean()

print("Married men (>50K):", married_high_income)
print("Single men (>50K):", single_high_income)
```

**9. What is the greatest number of hours a person works per week (*hours per week* attribute)? How many people work this number of hours, and what percentage of them earn a lot?**

```python
# Maximum hours worked
max_hours = data["hours-per-week"].max()

# People who work the maximum hours
max_workers = data[data["hours-per-week"] == max_hours]

# Number of such people
num_workers = max_workers.shape[0]

# Percentage earning >50K
high_income_percent = (max_workers["salary"] == ">50K").mean() * 100

print("Max hours per week:", max_hours)
print("Number of people working these hours:", num_workers)
print("Percentage earning >50K:", high_income_percent)
```

**10. Calculate the average working hours (*hours-per-week*) of low- and high-earners (*salary*) for each country (*native-country*).**

```python
avg_hours = data.groupby(["native-country", "salary"])["hours-per-week"].mean()

print(avg_hours)
```

**11. Group people by age groups: *young*, *adult*, *retiree*, where:** * *young* corresponds to ages 16-35 * *adult* corresponds to ages 35-70 * *retiree* corresponds to ages 70-100 **Enter the name of the corresponding age group for each person in the new AgeGroup column**

```python
bins = [16, 35, 70, 100]
labels = ["young", "adult", "retiree"]

data["AgeGroup"] = pd.cut(data["age"], bins=bins, labels=labels, right=False)

print(data[["age", "AgeGroup"]].head())
```

**12-13. Determine the number of people earning >50K in each age group (AgeGroup column), and also output the name of the age group in which people most often earn more than 50K (>50K)**

```python
# Create age groups
bins = [16, 35, 70, 100]
labels = ["young", "adult", "retiree"]
data["AgeGroup"] = pd.cut(data["age"], bins=bins, labels=labels, right=False)

# Count people earning >50K in each group
high_earners = data[data["salary"] == ">50K"].groupby("AgeGroup").size()

print(high_earners)

# Age group with the most high earners
most_high_earners = high_earners.idxmax()

print("Age group with most >50K earners:", most_high_earners)
```

**14. Group people by employment type (the occupation column) and determine the number of people in each group. Then write a filter function, filter_func, that will return only those groups in which the average age (the age column) is no more than 40 and in which all employees work more than 5 hours per week (the hours-per-week column).**

```python
# Number of people in each occupation
occupation_counts = data.groupby("occupation").size()

print(occupation_counts)
```

```python
# Create the filter function
def filter_func(group):
    return (group["age"].mean() <= 40) and (group["hours-per-week"].min() > 5)
```

```python
# Apply the filter
filtered_groups = data.groupby("occupation").filter(filter_func)

print(filtered_groups)
```
